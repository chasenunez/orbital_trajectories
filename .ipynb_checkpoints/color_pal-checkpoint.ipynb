```python
import matplotlib.pyplot as plt
import math

colors = [
    "#f38b8a", "#ffd393", "#cb2855", "#c4bd6c", "#d73423",
    "#706353", "#1f8a70", "#979330", "#e19222", "#979330",
    "#979330", "#979330", "#aec7b2", "#e6dfcf", "#e6dfcf",
    "#e6dfcf", "#e6dfcf", "#e6dfcf", "#e6dfcf", "#e6dfcf",
    "#e6dfcf", "#b6a965", "#d05227", "#7EBCE5", "#cb2655",
    "#e1d9c4", "#A73E5C", "#F34E52", "#738cc1"
]

labels = [
    "MBA", "TJN", "OMB", "GRK", "CEN",
    "MCA", "AST", "TNO", "AMO", "IMB",
    "APO", "ATE", "IEO", "HYA", "PAR",
    "JFc", "COM", "HYP", "HTC", "Etc",
    "JFC", "CTc", "Mercury", "Venus", "Earth",
    "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"
]

n = len(colors)
cols = 5
rows = math.ceil(n / cols)

fig, ax = plt.subplots(figsize=(cols * 3, rows * 2.2))
ax.set_xlim(0, cols)
ax.set_ylim(0, rows)
ax.axis("off")

for i, (color, label) in enumerate(zip(colors, labels)):
    row = rows - 1 - i // cols
    col = i % cols

    ax.add_patch(
        plt.Rectangle((col, row), 1, 1, facecolor=color, edgecolor="black")
    )
    ax.text(
        col + 0.5, row - 0.15, label,
        ha="center", va="top", fontsize=10
    )
    ax.text(
        col + 0.5, row + 0.5, color,
        ha="center", va="center", fontsize=8,
        color="black" if color.lower() != "#000000" else "white"
    )

plt.tight_layout()
plt.show()
```
